```python
cols = list(df.columns)
```

```python
def plot_trends_and_total(start, stop):
    fig, axes = plt.subplots(nrows=2, figsize=(12,10), sharex=True)
    df.iloc[start:stop].plot(y=cols[:-1], legend=False, linewidth=1, ax=axes[0])
    df.iloc[start:stop].plot(y=cols[-1], linewidth=1, ax=axes[1])
```

### A pretty sunny day and a partially cloudy day:

```python
plot_trends_and_total(31500, 32030)
```

### 6-hour close-up during partially cloudy day

```python
plot_trends_and_total(31900, 31972)
```

# Up-ramp event

```python
plot_trends_and_total(1125, 1160)
```

This is a close-up of a morning start-up period on a cloudy morning.

```python
plot_trends_and_total(10525, 10600)
```

### Autocorrelation of total power trend

```python
def plot_autocorrelation(start, finish, lags):
    fig = plt.figure(figsize=(12,8))
    ax1 = fig.add_subplot(211)
    fig = sm.graphics.tsa.plot_acf(df['total_power'].loc[start:finish].values.squeeze(), lags=lags, ax=ax1, fft=True)
    ax2 = fig.add_subplot(212)
    fig = sm.graphics.tsa.plot_pacf(df['total_power'].loc[start:finish], lags=lags, ax=ax2)
```

#### Autocorrelation function over lag of 6 hours with one month of data:

```python
plot_autocorrelation('2015-7-15', '2015-8-15', 6*12)
```

#### Autocorrelation function over lag of 6 hours on a sunny day

```python
df['total_power'].loc['2015-7-25':'2015-7-25'].plot()
```

```python
plot_autocorrelation('2015-7-25', '2015-7-25', 6*12)
```

#### Autocorrelation function over lag of 6 hours on a variable day

```python
df['total_power'].loc['2015-7-18':'2015-7-18'].plot()
```

```python
plot_autocorrelation('2015-7-18', '2015-7-18', 6*12)
```

Note that the autocorrelation function drops off much faster on the cloudy day than the sunny day, which matches intuition. On the sunny day, it takes 4.8 hours for the trend to become uncorrelated. On the cloudy day, it only takes three hours.
