```python
from core.preprocessing import StatisticalClearSky
from core.utilities import day_slice_from_date_range
import numpy as np
import pandas as pd
df = pd.read_pickle('data/master_dataset.pkl').fillna(0)
df = df.loc['2015-07-15':'2017-07-14']
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style='darkgrid', palette='colorblind', context='talk')
SUMMER = '2015-8-17'
FALL = '2015-11-17'
WINTER = '2016-2-16'
SPRING = '2016-5-2'
from numpy.linalg import norm
```

```python
cols = ['summer_mse', 'summer_bias', 'fall_mse', 'fall_bias', 'winter_mse', 'winter_bias', 'spring_mse', 'spring_bias',
        'sv_ratio']
analysis = pd.DataFrame(columns=cols, index=range(1,75))
summer_target = df.loc[SUMMER]
fall_target = df.loc[FALL]
winter_target = df.loc[WINTER]
spring_target = df.loc[SPRING]
summer_idx = day_slice_from_date_range(df.index, SUMMER)
fall_idx = day_slice_from_date_range(df.index, FALL)
winter_idx = day_slice_from_date_range(df.index, WINTER)
spring_idx = day_slice_from_date_range(df.index, SPRING)
for i in xrange(1, 75):
    if i != 74:
        key = 'S{:02}'.format(i)
    else:
        key = 'total_power'
    scs = StatisticalClearSky(df[key])
    scs.make_clearsky_model(n=5, plot=False)
    sv_ratio = scs.D[1] / scs.D[0]
    sum_cl = scs.estimate_clearsky(summer_idx[0][0])
    fal_cl = scs.estimate_clearsky(fall_idx[0][0])
    win_cl = scs.estimate_clearsky(winter_idx[0][0])
    spr_cl = scs.estimate_clearsky(spring_idx[0][0])
    r_sum = summer_target[key]/np.max(summer_target[key]) - sum_cl / np.max(summer_target[key])
    r_fal = fall_target[key]/np.max(fall_target[key]) - fal_cl / np.max(fall_target[key])
    r_win = winter_target[key]/np.max(winter_target[key]) - win_cl / np.max(winter_target[key])
    r_spr = spring_target[key]/np.max(spring_target[key]) - spr_cl / np.max(spring_target[key])
    sum_mse = norm(r_sum) / np.float(len(r_sum))
    sum_bias = np.average(r_sum)
    fal_mse = norm(r_fal) / np.float(len(r_fal))
    fal_bias = np.average(r_fal)
    win_mse = norm(r_win) / np.float(len(r_win))
    win_bias = np.average(r_win)
    spr_mse = norm(r_spr) / np.float(len(r_spr))
    spr_bias = np.average(r_spr)
    analysis.loc[i] = [sum_mse, sum_bias, fal_mse, fal_bias, win_mse, win_bias, spr_mse, spr_bias, sv_ratio]
    
```

```python
analysis.plot(y='sv_ratio')
```

```python
fig, axes = plt.subplots(nrows=2)
analysis.plot(y=analysis.columns[0], ax=axes[0])
analysis.plot(y=analysis.columns[1], ax=axes[1])
```
