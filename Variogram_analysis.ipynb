# Variogram analysis

In order to run this notebook, we have to grab the following packages:

```julia
for pkg in ["DataFrames","GeoStats","Plots","GR","Interact"]
    Pkg.add(pkg)
end

Pkg.checkout("GeoStats") # get the latest unstable version
```

```julia
using DataFrames

df = readtable("data/data.csv")
completecases!(df);
```

```julia
using GeoStats
using Plots, Interact

colnames = names(df)[2:21]

@manipulate for series1 in colnames, series2 in colnames
    yi = df[1:500,series1]
    yj = df[1:500,series2]
    X = collect(1:length(yi))'
    
    plot(EmpiricalVariogram(X, yi, yj, nbins=50), size=(1000,400))
end
```
